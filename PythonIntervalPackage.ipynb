```python
class interval:
    
    def __init__(self,left=0,right=0):
        if (left<=right):
            #Το κάνω round για να μην βγάζει εξοφρενικά μικρά αποτελέσματα.Λόγο της προσέγγισης που κάνει η newton
            self.left=round(left,10)
            self.right=round(right,10)
        else :
            print "Error : Wrong Interval"
    
    def __str__(self):
        return '[%s, %s]' % (self.left, self.right) 
    
    def __repr__(self):
        return '[%f , %f]' %(self.left, self.right)

    def __add__(self, rhs):
        if isinstance(rhs, interval):
            return interval(self.left + rhs.left, self.right + rhs.right)
        elif isinstance(rhs, int):
            return interval(self.left + rhs, self.right + rhs)
        elif isinstance(rhs, float):
            return interval(self.left + rhs, self.right + rhs)
        else:
            return NotImplemented
        
    def __sub__(self, rhs):
        if isinstance(rhs, interval):
            return interval(self.left - rhs.right, self.right - rhs.left)
        elif isinstance(rhs, int):
            return interval(self.left - rhs, self.right - rhs)
        elif isinstance(rhs, float):
            return interval(self.left - rhs, self.right - rhs)
        else:
            return NotImplemented
    
    def __mul__(self, rhs):
        if isinstance(rhs, interval):
            return interval(min(self.left*rhs.right,self.left*rhs.left,self.right*rhs.left,self.right*rhs.right),
                            max(self.left*rhs.right,self.left*rhs.left,self.right*rhs.left,self.right*rhs.right) ) 
        elif isinstance(rhs, int):
            return interval(min(self.left*rhs,self.right*rhs),
                            max(self.left*rhs,self.right*rhs) )             
        elif isinstance(rhs, float):
            return interval(min(self.left*rhs,self.right*rhs),
                            max(self.left*rhs,self.right*rhs) )              
        else:
            return NotImplemented
    
    def __div__ (self,rhs):
        if isinstance(rhs, interval):
            if rhs.contain(0):
                if (self.right>0 and self.left<0) or (self.right==0 and self.left==0) or(rhs.right==0 and rhs.left==0):
                    return interval(float("-inf"),(float("inf")))
                if (self.right <=0 and rhs.left<rhs.right and rhs.right == 0):
                    return interval(float (self.right)/rhs.left,(float("inf")))
                if (self.right <=0 and rhs.left<0 and 0<rhs.right):
                    temp=[interval(float("-inf"),float (self.right)/rhs.right),interval(float (self.right)/rhs.left,float("inf"))]
                    return temp
                if (self.right <=0 and rhs.left == 0 and rhs.right>rhs.left):
                    return interval((float("-inf"),float (self.right)/rhs.right))
                if (self.left >=0 and rhs.left < rhs.right and rhs.right==0):
                    return interval((float("-inf"),float (self.left)/rhs.left))
                if (self.left >=0 and rhs.left < 0 and rhs.right > 0):
                    temp=[interval(float("-inf"),float (self.left)/rhs.left),interval(float (self.left)/rhs.right,float("inf"))]
                    return temp
                if (self.left >=0 and rhs.left == 0 and rhs.right > rhs.left):
                    return interval(float (self.left)/rhs.right,(float("inf")))
            else :
                 return interval(min(self.left*(1./rhs.right),self.left*(1./rhs.left),self.right*(1./rhs.left),self.right*(1./rhs.right)),
                    max(self.left*(1./rhs.right),self.left*(1./rhs.left),self.right*(1./rhs.left),self.right*(1./rhs.right) ) )
        
        elif  isinstance(rhs, int):
            return interval(min(self.left*(1./rhs),self.right*(1./rhs)),
                            max(self.left*(1./rhs),self.right*(1./rhs) ) )
                                     
        elif  isinstance(rhs, float):
            return interval(min(self.left*(1./rhs),self.right*(1./rhs)),
                            max(self.left*(1./rhs),self.right*(1./rhs) ) )
        else :
            return NotImplemented
    
    def __pow__ (self,n):
        if n==0 :
            return interval(1,1)
        if (self.left>=0) or (self.left<=0 and 0 <=self.right and n%2==1):
            return interval(self.left**n,self.right**n)
        if (self.right <=0 and n%2==1):
            return interval(self.left**n,self.right**n)
        if (self.right <=0 and n%2==0):
            return interval(self.right**n,self.left**n)  
        if (self.left<=0 and 0 <=self.right and n%2==0):
            return interval(0,max(self.right**n,self.left**n))
            
    def __and__ (self,rhs):
        #Διάστημα με διάστημα.
        if isinstance(rhs, interval):
            if not(rhs.right<self.left or self.right<rhs.left):
                return interval(max(self.left,rhs.left),min(self.right,rhs.right))
        #Διάστημα με λίστα η οποία λίστα εκπροσωπεί την ένωση.
        if isinstance(rhs, list):
            for i in range (len(rhs)):
                if not(rhs[i].right<self.left or self.right<rhs[i].left):
                    return interval(max(self.left,rhs[i].left),min(self.right,rhs[i].right))
    
    def __or__ (self,rhs):
        #Διάστημα με διάστημα.
        if isinstance(rhs, interval):
            if not(rhs.right<self.left or self.right<rhs.left):
                return interval(min(self.left,rhs.left),max(self.right,rhs.right))
        #Διάστημα με λίστα η οποία λίστα εκπροσωπεί την ένωση.
        if isinstance(rhs, list):
            for i in range (len(rhs)):
                if not(rhs[i].right<self.left or self.right<rhs[i].left):
                    return interval(min(self.left,rhs[i].left),max(self.right,rhs[i].right))
                
    def __abs__(self):
        return max(abs(self.right),abs(self.left))
    
    def mignitude(self):
        return min(abs(self.right),abs(self.left))

    
    def sup(self):
        return self.right

    def inf(self):
        return self.left
    
    def midpoint(self):
        return float(self.right+self.left)/2
    
    def width(self):
        return float(self.right-self.left)
    
    def radius(self):
        return float(self.width())/2
    
    def subtract (self):
        mid=self.midpoint()
        x1=interval(self.left,mid)
        x2=interval(mid,self.right)
        return (x1,x2) 
    
    def contain (self , num):
            if self.left <= num and num <= self.right :
                return True
            else:
                return False

    def bisection(self,f,tol):
        L=[] 
        L.append(self) 
        Q=[] 
        k=0 
        while L:
            k=k+1
            self=L.pop()
            F=f(self)
            if F.contain(0):
                if abs(self.width())<tol:
                    Q.append(self)
                else :
                    x1,x2=self.subtract()
                    L.append(x1)
                    L.append(x2)
        print "Iterations :", k
        return Q  
    
    def NewtonStep(self,f,df):               
        mid=self.midpoint()
        ar=interval(f(mid),f(mid))
        ar2=ar/df(self)
        ar3=interval(mid,mid)-ar2
        return ar3

    def Newton(self,f,df,tol):
        L=[] 
        L.append(self) 
        Q=[] 
        k=0 
        while L:
            k=k+1
            self=L.pop() 
            F=f(self)
            dF=df(self) 
            if F.contain(0):
                if abs(self.width()) <tol:
                    Q.append(self) 
                elif  dF.contain(0):
                    x1,x2=self.subtract()
                    L.append(x1)
                    L.append(x2)
                else:
                    xnew=self.NewtonStep(f,df)
                    if xnew&self: 
                        L.append(xnew&self)
        print "Iterations :",k
        return Q
    
    def MRF(self,f,tol):
        Q=[] 
        L=[] 
        if f(self.left)==0:
            Q.append(self)
        else:
            L.append(self) 
        k=0 
        while L:
            k=k+1
            self=L.pop()
            Fdiast=f(self)
            if Fdiast.contain(0):
                if self.width()<tol:
                    Q.append(self)
                else :
                    x1,x2=self.subtract()
                    a1=x1.left
                    a2=x1.right
                    b1=x2.left
                    b2=x2.right
                    if (f(a1)*f(a2))<=0:
                        L.append(x1)
                    elif (f(b1)*f(b2))<=0:
                        L.append(x2)
                    else:
                        L.append(x1)
                        L.append(x2)
            Q.sort()
            for i in range(len(Q)):
                if i!=0:
                    del Q[i]
        print "Iterations :",k                               
        return Q

def hansen(f,sbox,w):
    sbox.sort()
    Lnew=[]
    n=0
#Β1: Θέσε Y=box X
    y=sbox
    
#Β2: c: Διάνυσμα μέσων 2 διαστημάτων και t(=fperisp) = άνω άκρο του διαστηματικού υπολογισμού της f στο c
    c=[] 
    c.append(y[0].midpoint())
    c.append(y[1].midpoint())
    t=f(c[0],c[1])

#B3: Έυρεση min F(Y) και θέσε y= min(F(Y))
    fy=f(sbox[0],sbox[1])
    ymin=fy.left

#B4: Αρχικοποίηση της λίστας
    L=[(y,ymin)]

#     while n<=iterr:
    while max(y[0].width(),y[1].width())>w:

#Β5: Διάλεξε μια συντεταγμένη διεύθυνσης k παράλληλη στην πλευρά του box Y η οποία έχει το μέγιστο μήκος 
#Β6: Διαίρεσε την με το κουτί Y ως προς την k σε 2 υποκουτιά V1,V2 (έστω v1=v2) που η ένωσή τους να δίνει το Y
        n=n+1
        v1=[]
        v2=[]
        if L[0][0][0].width()>L[0][0][1].width():
            k=L[0][0][0]
            mid=k.midpoint()
            v1.append(interval(k.left,mid))
            v1.append(L[0][0][1])
            v2.append(interval(mid,k.right))
            v2.append(L[0][0][1])

        else:
            k=L[0][0][1]
            mid=k.midpoint()
            v1.append(L[0][0][0])
            v1.append(interval(k.left,mid))
            v2.append(L[0][0][0])
            v2.append(interval(mid,k.right))

#Β7: fv: Περιέχει τα F(V1) και F(V2) που υπολογίζονται
        fv=[]
        fv.append(f(v1[0],v1[1]))
        fv.append(f(v2[0],v2[1]))

#Β8: vimin-> Κάτω άκρο της F(Vi) για i=1,2
        v1min=fv[0].left
        v2min=fv[1].left

#B9: Lnew -> νέα λιστα L η οποία περιέχει στο τέλος της τα ζευγάρια (Vi,vi) 
        Lnew=[]
        Lnew.append((v1,v1min))
        Lnew.append((v2,v2min))

#Β10: Διαλέγουμε το ζευγάρι της Lnew το οποίο να έχει το μικρότερο vimin(2o στοιχείο)
        k=float("inf")
        for i in range(len(Lnew)):
            if Lnew[i][1]< k:
                k=Lnew[i][1]
                pos=Lnew[i][0]

#Β11: Διαγραφή από τη λίστα κάθε (Ζ,z) όπου t<z
        Lcp=list(Lnew)
        for i in Lnew:
            if i[1] > t:
                Lcp.remove(i)
        Lnew=list(Lcp)
        L=list(Lnew)

#Β12: Το στοιχείο της λίστας που βρέθηκε στο Β10 τίθεται (Y,ymin) και c=mid(Y) και t=min(t,F(c))
        y=pos
        ymin=k
        c=[] 
        c.append(y[0].midpoint())
        c.append(y[1].midpoint())
        t=min(t,f(c[0],c[1]))

    return(y)
    
#     def mptest(L):
#     t=interval(float("-inf"),float("inf"))
#     L.sort()
#     Lcp=list(L)
#     hel=[]
#     for i in L:
#         hel.append((f(i),i))
#     hel.sort(key=itemgetter(0), reverse=True)
#     c=hel[0][1].midpoint[0].inf
#     t=min(f(c),t)
#     for i in L:
#         if f(i)[0].inf >= t:
#             Lcp.remove(i)
#     return (Lcp,len(L)-len(Lcp))
```
